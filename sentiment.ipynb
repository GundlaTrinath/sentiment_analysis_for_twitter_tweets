## Sentiment Analysis using LogisticRegression
##### dataset from kaggle of twitter tweets

```python
# import libraries
import pandas as pd
```

```python
# read excel data
df = pd.read_csv('dataset.csv', header=None, encoding='ISO-8859-1')
print(df.shape)
```

```python
# check for null values
df.isnull().sum()
```

```python
# filtering by columns
df = df.loc[:,[0,1,5]]
df.columns = ['target', 'id', 'text']
df
```

```python
# cleaning the text by using the Data preprocessing techniques
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stem = PorterStemmer()

def text_clean(text):
    text_without_links = re.sub(r'http\S+|www\S+', '', text)
    text_without_at_words = ' '.join([word for word in text_without_links.split() if not word.startswith('@')])
    alpha_text = re.sub(r'[^A-Za-z\s]', '', text_without_at_words)
    cleaned_text = ' '.join([stem.stem(word) for word in alpha_text.split() if word not in stopwords.words('english')])
    
    return cleaned_text

df['text'] = df['text'].apply(text_clean)
df
```

```python
df = df[['target','text']]
df.head()
```

```python
# import libraries for model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# spliting into train and test
text_data=df['text'].values.tolist()
target_data = df['target'].values.tolist()
X_train, X_test, y_train, y_test = train_test_split(text_data, target_data, test_size=0.3, stratify=target_data, random_state=3)
```

```python
# converting to vectors 
tfidf = TfidfVectorizer()
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
# model creation
model = LogisticRegression(max_iter=100)
```

```python
# traning the model 
model.fit(X_train, y_train)
```

```python
# accuracy score 
X_test_prediction = model.predict(X_test)
score = accuracy_score(y_test, X_test_prediction)
score
```
